Opis implementacije
--------------
Učitavanje potrebnih biblioteka:

U prvom koraku, uvozimo potrebne Python biblioteke. pandas koristimo za manipulaciju podacima u DataFrame-ovima, sqlalchemy za rad s SQLite bazom podataka, Flask za izradu web aplikacije s REST API-jem te matplotlib.pyplot za stvaranje grafova. Potom za učitavanje podataka iz JSON i CSV datoteka koristimo funkciju pd.read_csv() za učitavanje podataka iz CSV datoteke i funkciju pd.read_json() za učitavanje podataka iz JSON datoteke. Ova dva koraka su ključna za inicijalno stvaranje DataFrame-ova. Nakon učitavanja podataka i njihove obrade u zasebnim DataFrame-ovima, koristimo pd.concat() funkciju za spajanje DataFrame-ova po zajedničkim stupccima. To rezultira stvaranjem novog DataFrame-a koji kombinira informacije iz oba izvorna DataFrame-a.Slijedeća na redu je pohrana podataka u bazu podataka kako bi ih imali spremljene. Uspostavljamo vezu s SQLite bazom pomoću 
db_url = 'sqlite:///baza.db'
engine = create_engine(db_url). 
Zatim koristimo to_sql() metodu DataFrame-a kako bismo pohranili kombinirane podatke u SQLite bazu. Ovdje pretpostavljamo da veza s bazom podataka već postoji. Nakon što smo podatke sigurno pohranili, te ih imamo spremljenje kako nebi došlo do gubitka koristimo Flask, mikro web framework za Python, kako bismo implementirali REST API. Definiramo jednostavnu web aplikaciju s jednim endpointom koji iz baze dohvaća podatke. Ovaj dio sustava omogućuje komunikaciju s podacima putem HTTP zahtjeva. Aplikacija se nalazi u posebnoj datoteci. Te na samom kraju koristimo matplotlib.pyplot za stvaranje grafova temeljenih na podacima iz kombiniranog DataFrame-a kako bi mogli napraviti analizu podataka. Ovisno o našim potrebama, mogu se generirati različite vrste grafova (linijske, stupčaste, kružne itd.) kako bismo vizualizirali podatke na željeni način.

#### Pretprocesiranje i integracija podataka iz više skupova

```python
columns_to_drop = ['address', 'region']
columns_to_drop = [col for col in columns_to_drop if col in df.columns]

if columns_to_drop:
    df = df.drop(columns=columns_to_drop)
df.head(5)
```

```python
df_kombinirano = pd.concat([df, dfj], ignore_index=True)
df_kombinirano
```

```python
df_kombinirano['started_using'] = pd.to_datetime(df_kombinirano['started_using'])
max_datum = df_kombinirano['started_using'].max()
df_kombinirano['ukupno_vrijeme_koristenja'] = max_datum - df_kombinirano['started_using']
df_kombinirano
```

```python
df_kombinirano = df_kombinirano.rename(columns={
    'name': 'ime',
    'phone': 'broj_mobitela',
    'email': 'e-mail',
    'country': 'drzava',
    'started_using': 'pocetak_koristenja',
    'subscribed': 'pretplacen',
    'age': 'starost'
})
df_kombinirano
```

```python
nevaljani_datumi = df_kombinirano['pocetak_koristenja'][pd.to_datetime(df_kombinirano['pocetak_koristenja'], errors='coerce').isnull()]

if not nevaljani_datumi.empty:
    print(f"Redovi s neispravnim datumima:\n{df_kombinirano.loc[nevaljani_datumi.index]}")
else:
    print("Nema neispravnih datuma.")
```

```python
prazne_celije = df_kombinirano[df_kombinirano.isnull().any(axis=1)]

if not prazne_celije.empty:
    print(f"Redovi s praznim ćelijama:\n{prazne_celije}")
else:
    print("Nema praznih ćelija.")
```

```python
jed_vri = df_kombinirano['pretplacen'].unique()
print("Jedinstvene vrijednosti u 'pretplacen':", jed_vri)

df_kombinirano['pretplacen'] = df_kombinirano['pretplacen'].map({'Yes': 1, 'No': 0})

df_kombinirano
```

#### Način pohrane integriranog skupa podataka

```python
db_url = 'sqlite:///spotify.db'
engine = create_engine(db_url)

df_kombinirano['ukupno_vrijeme_koristenja'] = df_kombinirano['ukupno_vrijeme_koristenja'].dt.total_seconds()
df_kombinirano.to_sql('podaci', engine, index=False, if_exists='replace')

print("Podaci pohranjeni u bazu.")
```
